```python
import pandas as pd
```

```python
df_area = pd.read_csv('data/area.csv')
df_crime = pd.read_csv('data/crime.csv')
df_premises = pd.read_csv('data/premises.csv')
df_status = pd.read_csv('data/status.csv')
df_weapon = pd.read_csv('data/weapon.csv')
df_mocodes = pd.read_csv('data/mocodes.csv')
```

### Area

```python
df_area.head()
```

```python
df_area.isna().sum()
```

```python
df_area.duplicated(subset='area_code').sum()
```

```python
df_area.location.sample(10)
```

```python
df_area = (df_area
           .drop(columns=['location'])
           .drop_duplicates()
           .sort_values('area_code')
           .reset_index(drop=True))
df_area
```

### Crime

```python
df_crime 
```

```python
df_crime.isna().sum()
```

```python
df_crime = df_crime.sort_values('crime_code').reset_index(drop=True)
df_crime
```

### Premises

```python
df_premises
```

```python
df_premises.isna().sum()
```

```python
df_premises = df_premises.dropna(subset=['premises_code'])
df_premises.isna().sum()
```

```python
df_premises[df_premises.premises_description.isna()]
```

```python
df_premises.loc[df_premises.premises_description.isna(), 'premises_description'] = 'DESCRIPTION NOT AVAILABLE'
df_premises.isna().sum()
```

```python
df_premises = df_premises.sort_values('premises_code').reset_index(drop=True)
df_premises
```

### Status

```python
df_status
```

```python
df_status = df_status.dropna()
df_status
```

```python
df_status = df_status.replace({'UNK': 'UNKNOWN'})
df_status
```

### Weapon

```python
df_weapon
```

```python
df_weapon.isna().sum()
```

```python
df_weapon = df_weapon.dropna().drop_duplicates().sort_values('weapon_code').reset_index(drop=True)
df_weapon
```

```python
df_weapon.weapon_code = df_weapon.weapon_code.astype(int)
df_weapon
```

### Mo Codes

```python
df_mocodes
```

```python
df_mocodes.isna().sum()
```

```python
df_area.to_csv('data/area.csv', index=False)
df_crime.to_csv('data/crime.csv', index=False)
df_premises.to_csv('data/premises.csv', index=False)
df_status.to_csv('data/status.csv', index=False)
df_weapon.to_csv('data/weapon.csv', index=False)
df_mocodes.to_csv('data/mocodes.csv', index=False)
```
