```python
import pandas as pd
from utils import show_num_ranges, show_cat_values
```

```python
df = pd.read_pickle('data/event.pkl')
df.head()
```

```python
df.shape
```

```python
df = df.explode('crime_codes', ignore_index=True).dropna(subset=['crime_codes'])
df.head()
```

```python
df.shape
```

```python
df = df.explode('mocodes', ignore_index=True).dropna(subset=['mocodes'])
df.head()
```

```python
df.shape
```

```python
show_num_ranges(df)
```

```python
df.victim_age = df.victim_age.astype('uint8')
df.premises_code = df.premises_code.astype('category')
df.weapon_code = df.weapon_code.astype('category')
show_num_ranges(df)
```

```python
show_cat_values(df)
```

```python
df.crime_codes = df.crime_codes.astype('category')
df.mocodes = df.mocodes.astype('category')
df.victim_sex = df.victim_sex.astype('bool')
show_cat_values(df)
```

```python
df.to_pickle('data/event.pkl')
```
